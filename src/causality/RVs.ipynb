```python
import numpy as np
```

```python
# two independent RVs

theta = 0.9
v = 0.4

n_samples = 1000
x = np.random.choice(2, p = [1 - theta, theta], size = n_samples)
y = np.random.choice(2, p = [1 - v, v], size = n_samples)
```

```python
np.mean(x), np.mean(y)
```

```python
np.mean(x[y==0]), np.mean(x[y==1]) # xs where y is 0 versus 1
```

```python
# two dependent discrete RVs
theta = 0.9
n_samples = 100000
x = np.random.choice(2, p = [1 - theta, theta], size = n_samples)

v = np.zeros(2)
v[0] = 0.2
v[1] = 0.6
y = np.array([np.random.choice(2, p = [1 - v[x_t], v[x_t]]) for x_t in x]) # make the list into an array
```

```python
np.mean(x), np.mean(y)
```

```python
np.mean(x[y==0]), np.mean(x[y==1]) # xs where y is 0 versus 1
```

```python
np.mean(y[x==0]), np.mean(y[x==1]) # xs where y is 0 versus 1
```

```python
# Continuous x, y: independent
n_samples = 1000
x = np.random.normal(0, 1, size = n_samples)
y = np.random.normal(1,1, size = n_samples)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
# Continuous x, y: dependent
n_samples = 1000
x = np.random.normal(0, 1, size = n_samples)
y = np.random.normal(x, 1, size = n_samples) # x is an input to y
plt.scatter(x,y)
```

```python
y = np.random.normal(0, 1, size = n_samples)
x = np.random.normal(y, 1, size = n_samples) # y is an input to x
plt.scatter(x,y)
```

```python
# Discrete y, continuous x

y = np.random.choice(2, size = n_samples)
x = np.random.normal(0, 1, size = n_samples)
plt.hist(x[y==0], alpha =0.5)
plt.hist(x[y==1], alpha=0.5)
```

```python
# Discrete y, continuous x - dependence

y = np.random.choice(2, size = n_samples)
x = np.random.normal(y, 1, size = n_samples)
plt.hist(x[y==0], alpha =0.5)
plt.hist(x[y==1], alpha=0.5)
```
