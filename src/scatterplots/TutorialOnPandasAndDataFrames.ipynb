```python
# tables in numpy
import numpy as np
n_data = 1000
n_features = 4
data = np.random.uniform(size=[n_data, n_features])

# data slices
data[0:3,:] # the first three lines
data[:,1] # the second column
data[:,[1,3]] # the first and third column
```

```python
plt.scatter(data[:, 1], data[:,2])
```

```python
plt.hist2d(data[:, 1], data[:,2])
```

```python
import pandas as pd
df = pd.DataFrame(data, columns = ["Alcohol", "Caffeine", "Sugar", "THC"])
```

```python
# data slices
data[0:3,:] # the first three lines
data[:,1] # the second column
data[:,[1,3]] # the first and third column
df["Alcohol"] # pick the column "Alcohol in the data frame"
df.loc[:,"Alcohol"] #this is equivalent to the above
df.loc[:, ["Alcohol", "Sugar"]]
```
